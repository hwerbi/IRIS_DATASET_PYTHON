```python
import matplotlib
import pylab as pl
pl.style.use('ggplot')

from sklearn import datasets
from itertools import cycle
```

```python
# function to plot the data 
# each time a different axis couple is used for the display
def plot_2D(data, target, target_names,v1,v2):
 colors = cycle('rgbcmykw') 
 target_ids = range(len(target_names))
 pl.figure()  
 for i, c, label in zip(target_ids, colors, target_names):
     pl.scatter(data[target == i, v2], data[target == i, v1], c=c, label=label)
 
```

```python
#prepare the parameters of the function
irisData = datasets.load_iris()
target_names=['setosa', 'versicolor', 'virginica']
```

```python
plot_2D(irisData.data, irisData.target,target_names,0,2) 
pl.xlabel('sepal length (cm)')
pl.ylabel('petal length (cm)')
pl.title('Iris dataset: petal length vs sepal length')
pl.legend(loc='lower right')
pl.show()
```

```python
plot_2D(irisData.data, irisData.target,target_names,0,1) 
pl.xlabel('sepal length (cm)')
pl.ylabel('sepal width (cm)') 
pl.title('Iris dataset: sepal length vs sepal width')
pl.legend(loc='lower right')
pl.show()
```

```python
plot_2D(irisData.data, irisData.target,target_names,0,3) 
pl.xlabel('sepal length (cm)')
pl.ylabel('petal width (cm)') 
pl.title('Iris dataset: sepal length vs petal width')
pl.legend(loc='lower right')
pl.show() 
```

```python
plot_2D(irisData.data, irisData.target,target_names,1,2) 
pl.xlabel('sepal width (cm)')
pl.ylabel('petal length (cm)') 
pl.title('Iris dataset: sepal width vs petal length')
pl.legend(loc='lower right')
pl.show() 
```
